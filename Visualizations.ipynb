# IMPORT THE DATA:

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
mt_cars=pd.read_csv("mtcars.csv")
mt_cars.head(10)      
```

# EDA PART:

```python
x=mt_cars['mpg']
y=mt_cars['cyl']
```

```python
plt.pie(x=mt_cars["cyl"].value_counts(),autopct='%1.0f%%',labels=[8,4,6],explode=[0.02,0,0.03])
plt.show()
```

# UNIVARIATE ANALYSIS FOR CONTINUES DATA:

```python
#SIZE OF THE CHART:
plt.figure(figsize=(10,5))
#GIVING DATA AND VALUE FOR GRAPH:
plt.hist(x="mpg",data=mt_cars)
#LABELING X-AXIS
plt.xlabel("MPG")
#LABELING Y-AXIS
plt.ylabel("Frequency")
#GIVING TITLE:
plt.title("MPG Distribution",size=20)
#WILL SHOW ZOOMOUT:
plt.show()
```

```python
plt.figure(figsize=(5,5))
plt.hist(x="disp",data=mt_cars)
plt.xlabel("disp")
plt.ylabel("Frequency")
plt.title("Disp Distribution",size=20)
plt.show()
```

```python
print('BOXPLOT USING MATPLOTLIB')
plt.boxplot(x="mpg",data=mt_cars)
plt.show()

print('BOXPLOT USING SEABORN')
sns.boxplot(x=mt_cars['drat'])
plt.show()
```

# UNIVARIATE ANALYSIS FOR DISCRETE DATA:

```python
sns.countplot(x="cyl",data=mt_cars)
```

```python
plt.pie(x=mt_cars["cyl"].value_counts(),autopct='%1.0f%%',
        labels=[8,4,6],explode=[0.02,0,0.03])
plt.show()
```
